```python
import os
import xml.etree.ElementTree as ET
```

```python
def read_xml(xml_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()
    return root

def fix_folder_and_filename(root, folder, filename):
    annt = root.find('annotation')
    fldr = annt.find('folder')
    fldr.text = folder
    flnm = annt.find('filename')
    flnm.text = filename
     
```

```python
import cv2

img_folder = "../data/d1set11m/images"
xml_folder = "../data/d1set11m/xmls"
new_img_folder = "../data/d1set11m/new_images"
if not os.path.isdir(new_img_folder):
    os.mkdir(new_img_folder)
```

```python
imgs = os.listdir(img_folder)

for img in imgs:
    print(img)
    src = cv2.imread(os.path.join(img_folder, img))
    cv2.imwrite(os.path.join(new_img_folder, img[:-3] + 'jpg'), src)
```

```python
img_folder = "../data/d1set11m/images"
xml_folder = "../data/d1set11m/xmls"

imgs = sorted(os.listdir(img_folder))
xmls = sorted(os.listdir(xml_folder))
for img, xml in zip(imgs, xmls):
    print(img)
    tree = ET.parse(os.path.join(xml_folder, xml))
    root = tree.getroot()
    fldr = root.find("folder")
    fldr.text = img_folder
    flnm = root.find("filename")
    flnm.text = img
    if (obj := root.find("object")):
        if obj.find("pose") is None:
            pose = ET.SubElement(obj, "pose")
            pose.text = "Unspecified"
        for bbox in obj.findall("bndbox"):
            xmin = bbox.find("xmin")
            ymin = bbox.find("ymin")
            xmax = bbox.find("xmax")
            ymax = bbox.find("ymax")
            xmin.text = xmin.text.split(".")[0]
            ymin.text = ymin.text.split(".")[0]
            xmax.text = xmax.text.split(".")[0]
            ymax.text = ymax.text.split(".")[0]
    tree.write(os.path.join(xml_folder, xml))
```
