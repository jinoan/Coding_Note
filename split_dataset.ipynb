```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
root = '/home/noah/Desktop/SPC/data/full-name2/'
images_path = 'images/'
xmls_path = 'annotations/'
train_path = 'train/'
test_path = 'test/'
```

```python
if not os.path.isdir(root+train_path):
    os.mkdir(root+train_path)
if not os.path.isdir(root+train_path+'images/'):
    os.mkdir(root+train_path+'images/')
if not os.path.isdir(root+train_path+'xmls/'):
    os.mkdir(root+train_path+'xmls/')
if not os.path.isdir(root+test_path):
    os.mkdir(root+test_path)
if not os.path.isdir(root+test_path+'images/'):
    os.mkdir(root+test_path+'images/')
if not os.path.isdir(root+test_path+'xmls/'):
    os.mkdir(root+test_path+'xmls/')
```

```python
dnames = list(map(lambda x: x.strip('.jpg'), os.listdir(root+images_path)))
trains, tests = train_test_split(dnames, test_size=1100)
print(f'train len: {len(trains)} / test len: {len(tests)}')
```

```python
for i, train in enumerate(trains):
    print(f'train {i} - {train}.jpg')
    shutil.copy2(root+images_path+train+'.jpg', root+train_path+'images/'+train+'.jpg')
    shutil.copy2(root+xmls_path+train+'.xml', root+train_path+'xmls/'+train+'.xml')

for i, test in enumerate(tests):
    print(f'test {i} - {test}.jpg')
    shutil.copy2(root+images_path+test+'.jpg', root+test_path+'images/'+test+'.jpg')
    shutil.copy2(root+xmls_path+test+'.xml', root+test_path+'xmls/'+test+'.xml')
```
